```python
import cv2
import numpy as np
from rotateUtil import *
import math
import glob
import os
```

```python
ar = []
```

```python
path = "Anomalous/images"
for filename in glob.glob(os.path.join(path, '*.tif')):
    img = cv2.imread(filename, 0)
    ar.append(img)
    for i in range(60, 241, 60):
        a = np.concatenate((img[:, i+1:], img[:, 0:i+1]), axis=1)
        b = np.concatenate((img[:, img.shape[1]-i:], img[:, 0:img.shape[1]-i]), axis=1)
        ar.append(a)
        ar.append(b)
    imrot = rotate_image(img, 90)
    for i in range(60, 181, 60):
        a = np.concatenate((imrot[:, i+1:], imrot[:, 0:i+1]), axis=1)
        b = np.concatenate((imrot[:, imrot.shape[1]-i:], imrot[:, 0:imrot.shape[1]-i]), axis=1)
        ar.append(rotate_image(a, 270))
        ar.append(rotate_image(b, 270))
print len(ar)
```

```python
name = 1
imN = 10
rot = []
for image in ar:
    for deg in range(10, 351, 10):
        image_rotated = rotate_image(image, deg)
        image_height, image_width = image.shape[0:2]
        image_rotated_cropped = crop_around_center(
            image_rotated,
            *largest_rotated_rect(
                image_width,
                image_height,
                math.radians(deg)
            )
        )
        rot.append(cv2.resize(image_rotated_cropped, (32, 32), cv2.INTER_AREA))
        #cv2.imwrite("ana/"+str(name)+".jpg", cv2.resize(image_rotated_cropped, (100, 100), cv2.INTER_AREA))
        name = name + 1
        if(name%50==0):
            print "image: "+str(imN)+" name: "+str(name)
    if(imN%20==0):
        print "image: "+str(imN)
    imN = imN + 1
```

```python
import pickle
f = open("anmo.pickle", 'w')
pickle.dump(rot, f)
f.close()
```
